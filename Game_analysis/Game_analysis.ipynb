# Портрет пользователя каждого региона

__Функция построения графика-таблетки__

Принимает таблицу, сгруппированную по столбцу `category` и суммированием по столбцу `sales`

Возвращает график с соотношением продаж в регионе в разрезе платформ

```python
def create_pie(table):
    fig = plt.figure(figsize=(7,7))
    
#     table.groupby('category')['ratio'].sum().sort_values(ascending=False).plot(
    table.plot(
        kind='pie',
        title="Продажи в регионе",
        legend=True,
        autopct='%1.0f%%',
        shadow=True,
#         ylabel='Соотношение'
    )
    
    plt.ylabel('Соотношение')
```

__Функция по построению сводной таблицы__

Принимает значения:
- колонка_1, по значениям которой будем считать кол-во и сумму
- колонка_2 (индекс), по значениям которой будем группировать данные

Функция добавляет: 
- столбец `ratio` - соотношение суммы по группе в общей сумме значений
- столбец `category` - ТОП-5 возвращает имя платформы, остальным платформам присваивает категорию _others_


Функция возвращает:
- первые 6 строк таблицы, сгруппированной по колонке_2 и отсортированной по столбцу `ratio`
- выполнение функции _create_pie( )_

```python
def ratio (column_1, column_2):
    grouped = data_period.pivot_table(
        index= column_2.name,
        values= column_1.name,
        aggfunc=['count','sum']
    ).reset_index()
    
    grouped['ratio'] = grouped['sum'] / grouped['sum'].sum()
    grouped.sort_values(by='ratio', ascending=False, inplace=True)
    
    grouped['category'] = grouped[column_2.name].where(
        grouped[column_2.name].isin(
            grouped[column_2.name].head()), 'others')
    
    data_period['category'] = data_period[column_2.name].where(      # добавляем столбец с категорией в основную 
        data_period[column_2.name].isin(                             # таблицу
            grouped[column_2.name].head()), 'others')
    
    table_pie = data_period.groupby('category')[column_1.name].sum() # группируем таблицу для функции create_pie

    
    print(grouped.head(6))
    return create_pie(table_pie)
```

# Проверка гипотез

## Средние пользовательские рейтинги платформ _Xbox One_ и _PC_

Нулева гипотеза утверждает о равенстве среднего значения пользовательского рейтинга на платформах _Xbox One_ и _PC_, альтернативная гипотеза свидетельствует об обратном.

Н0: средний пользовательский рейтинг _Xbox One_ = средний пользовательский рейтинг _PC_

Н1: средние пользовательские рейтинги _Xbox One_ и _PC_ не равны

Р-уровень значимости = 0.05

Формируем срезы данных по каждой платформе

```python
xbox_raiting = data_period.loc[(data_period['platform'] == "XOne") & (data_period['user_score'] != 99)]

xbox_raiting.head(2)
```

```python
pc_raiting = data_period.loc[(data_period['platform'] == "PC") & (data_period['user_score'] != 99)]

pc_raiting.head(2)
```

Рассчитываем дисперсию для каждой платформы для дальнейшего выбора равенства дисперсий в Т-тесте

```python
def exploration(table_1, table_2):
    print(f'Дисперсия выборки-1:', np.var(table_1['user_score'], ddof=1))
    print(f'Дисперсия выборки-2:',np.var(table_2['user_score'], ddof=1))
    print()
    
    table_1['user_score'].plot(
        kind='kde', grid=True, title='Распределение средних значений', figsize=(10,5)
    )
    table_2['user_score'].plot(
        kind='kde', grid=True, title='Распределение средних значений', figsize=(10,5)
    )
    
    print(table_1['user_score'].describe())
    print()
    print(table_2['user_score'].describe())
```

```python
exploration(xbox_raiting, pc_raiting)
```

Сравниваем средние двух выборок с установленным уровнем значимости 0.05

```python
alpha = 0.05
results_1 = st.ttest_ind(xbox_raiting['user_score'], pc_raiting['user_score'], equal_var=False)
print(f'p-значение: {results_1.pvalue}')

if results_1.pvalue < alpha:
    print(f"Отвергаем Н0")
else:
    print(f'Не получилось отвергнуть Н0')
```

Для проверки гипотезы о равенстве средних значений двух выборок изпользовали тест Стьюдента. Мы не смогли отвергнуть нулевую гипотезу, так как p-значение значительно больше уровня значимости. Это означает, что вероятность того, что средние значения пользовательского рейтинга на платформах _Xbox One_ и _PC_ равны велика
